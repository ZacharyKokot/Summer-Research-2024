Bloch sphere for the example state $\ket{\psi} = \frac{1}{\sqrt{2}}(\ket{0} + \ket{1})$

```python
import numpy as np
import qutip

example_state = qutip.Qobj([1/np.sqrt(2),1/np.sqrt(2)])
bloch =  qutip.Bloch()

bloch.add_states(state=example_state, kind='vector')
bloch.vector_color = ['#af5afa']
bloch.show()
```
